```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = {
    "Normal": [10, 12, 11, 13, 12, 11, 10, 12, 13, 11],
    "Right_Skewed": [5, 6, 7, 8, 50, 60, 70, 80, 90, 100],
    "Left_Skewed": [100, 90, 80, 70, 60, 55, 50, 49, 48, 47]
}

df = pd.DataFrame(data)

plt.figure(figsize=(15, 5))

for i, col in enumerate(df.columns, 1):
    plt.subplot(1, 3, i)
    sns.histplot(df[col], kde=True, bins=10, color="skyblue")
    plt.title(f"{col}\nSkew = {df[col].skew():.2f}")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))

# Original
plt.subplot(1, 3, 1)
sns.histplot(df["Right_Skewed"], kde=True, color="salmon")
plt.title(f"Original\nSkew={df['Right_Skewed'].skew():.2f}")

# Log Transform
plt.subplot(1, 3, 2)
sns.histplot(df["Right_Skewed_log"], kde=True, color="skyblue")
plt.title(f"Log Transform\nSkew={df['Right_Skewed_log'].skew():.2f}")

plt.subplot(1, 3, 3)
sns.histplot(df["Right_Skewed_sqrt"], kde=True, color="lightgreen")
plt.title(f"Sqrt Transform\nSkew={df['Right_Skewed_sqrt'].skew():.2f}")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Example skewed data
np.random.seed(42)
df = pd.DataFrame({
    "Right_Skewed": np.random.exponential(scale=2, size=50)
})

# Transformations
df["Right_Skewed_log"] = np.log1p(df["Right_Skewed"])
df["Right_Skewed_sqrt"] = np.sqrt(df["Right_Skewed"])

# Line chart
plt.figure(figsize=(12,6))

plt.plot(df.index, df["Right_Skewed"], label="Original (Skewed)", marker="o")
plt.plot(df.index, df["Right_Skewed_log"], label="Log Transform", marker="s")
plt.plot(df.index, df["Right_Skewed_sqrt"], label="Sqrt Transform", marker="^")

plt.xlabel("Index")
plt.ylabel("Values")
plt.title("Line Chart of Original vs Transformed Data")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Example skewed data
np.random.seed(42)
df = pd.DataFrame({
    "Right_Skewed": np.random.exponential(scale=2, size=20)
})

# Transformations
df["Right_Skewed_log"] = np.log1p(df["Right_Skewed"])
df["Right_Skewed_sqrt"] = np.sqrt(df["Right_Skewed"])

plt.figure(figsize=(14,8))

# 1️⃣ Line color, style, marker
plt.plot(df.index, df["Right_Skewed"], 
         color="blue", linestyle="-", marker="o", linewidth=2, label="Original")

# 2️⃣ Dashed line with square markers
plt.plot(df.index, df["Right_Skewed_log"], 
         color="red", linestyle="--", marker="s", linewidth=2, label="Log Transform")

# 3️⃣ Dotted line with triangle markers
plt.plot(df.index, df["Right_Skewed_sqrt"], 
         color="green", linestyle=":", marker="^", linewidth=2, label="Sqrt Transform")

# Extra parameters
plt.title("Different Line Chart Parameters", fontsize=16, fontweight="bold")
plt.xlabel("Index", fontsize=12)
plt.ylabel("Values", fontsize=12)
plt.legend(loc="upper left", fontsize=10)
plt.grid(True, linestyle="--", alpha=0.7)
plt.axhline(y=df["Right_Skewed"].mean(), color="gray", linestyle="-.", label="Mean Line")

plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Example data
data = {
    "Category": ["A", "B", "C", "D"],
    "Product1": [10, 20, 15, 25],
    "Product2": [12, 18, 22, 28],
    "Product3": [8, 15, 19, 23]
}

df = pd.DataFrame(data)

# Bar width set karna
bar_width = 0.25
x = range(len(df["Category"]))  # X positions

# Plot bars for each product
plt.bar([i - bar_width for i in x], df["Product1"], width=bar_width, label="Product1", color="skyblue")
plt.bar(x, df["Product2"], width=bar_width, label="Product2", color="orange")
plt.bar([i + bar_width for i in x], df["Product3"], width=bar_width, label="Product3", color="green")

# Labels and title
plt.xticks(x, df["Category"])
plt.title("Comparison of 3 Products", fontsize=14, fontweight="bold")
plt.xlabel("Category", fontsize=12)
plt.ylabel("Values", fontsize=12)
plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Example data
data = {
    "Month": ["Jan", "Feb", "Mar", "Apr", "May"],
    "Product1": [100, 120, 90, 140, 160],
    "Product2": [80, 95, 110, 130, 150],
    "Product3": [60, 85, 105, 120, 140]
}

df = pd.DataFrame(data)

# --- Create Subplots ---
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# 1️⃣ Grouped Bar Chart
bar_width = 0.25
x = range(len(df["Month"]))
axes[0].bar([i - bar_width for i in x], df["Product1"], width=bar_width, label="Product1", color="skyblue")
axes[0].bar(x, df["Product2"], width=bar_width, label="Product2", color="orange")
axes[0].bar([i + bar_width for i in x], df["Product3"], width=bar_width, label="Product3", color="green")
axes[0].set_xticks(x)
axes[0].set_xticklabels(df["Month"])
axes[0].set_title("Grouped Bar Chart")
axes[0].legend()

# 2️⃣ Stacked Bar Chart
df.set_index("Month")[["Product1", "Product2", "Product3"]].plot(
    kind="bar", stacked=True, ax=axes[1], color=["skyblue", "orange", "green"]
)
axes[1].set_title("Stacked Bar Chart")

# 3️⃣ Horizontal Bar Chart
df.set_index("Month")[["Product1", "Product2", "Product3"]].plot(
    kind="barh", ax=axes[2], color=["skyblue", "orange", "green"]
)
axes[2].set_title("Horizontal Bar Chart")

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Example data: Random sales values
np.random.seed(42)
sales = np.random.randint(50, 200, 100)  # 100 random values between 50–200

df = pd.DataFrame({"Sales": sales})

# --- Histogram ---
plt.figure(figsize=(8,5))
plt.hist(df["Sales"], bins=10, color="skyblue", edgecolor="black")  # 10 bins (intervals)
plt.title("Sales Distribution Histogram")
plt.xlabel("Sales Value Range")
plt.ylabel("Frequency")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```
