```python
import pandas as pd

data = {
    "Category": ["Electronics", "Electronics", "Kitchen", "Kitchen", "Clothing", "Clothing"],
    "Brand": ["Sony", "Samsung", "Philips", "Panasonic", "Nike", "Adidas"],
    "Price": [1000, 1200, 300, 400, 150, 200],
    "Stock": [50, 60, 100, 80, 200, 150]
}

df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)

pivot = pd.pivot_table(
    df,
    values=["Price", "Stock"],   
    index="Category",            
    aggfunc={"Price": "mean", "Stock": "sum"} 
)

print("Pivot Table:")
print(pivot)
```

```python
melted = pivot.reset_index().melt(id_vars="Category", var_name="Brand", value_name="AvgPrice")

print("Melted DataFrame:")
print(melted)
```

```python
grouped = df.groupby("Category").agg(
    Avg_Price=("Price", "mean"),
    Total_Stock=("Stock", "sum")
).reset_index()

print("GroupBy Result:")
print(grouped)
```
