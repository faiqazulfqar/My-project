# Rows and Columns selection

```python
import pandas as pd

df = pd.read_csv(r"C:\Users\Gateway\Downloads\products-10000.csv")

selected_cols = df[['Name', 'Brand', 'Price', 'Stock']]

print("Selected columns:")
print(selected_cols.head())
```

```python
print("10th row using iloc:")
print(df.iloc[9])

print("\nRow with index = 10 using loc:")
print(df.loc[10])
```

```python
result = df.loc[0:9, ['Category', 'Price']]
print(result)
```

```python
result = df.loc[df['Category'] == "Kitchen Appliances", ['Name', 'Price']]
print(result)
```

# Filtering with Conditions

```python
result = df[df['Price'] > 500]

print(result.head())  
```

```python
out_of_stock = df[df['Stock'] == 0]

print("Products with Stock = 0:")
print(out_of_stock.head())   
```

```python
mueller_products = df[df['Brand'] == "Mueller Inc"]

print("Products from Mueller Inc:")
print(mueller_products.head()) 
```

```python
filtered_products = df[(df['Price'] > 300) & (df['Stock'] < 100)]

print("Products with Price > 300 and Stock < 100:")
print(filtered_products.head())   
```

```python
filtered_products = df[(df['Category'] == "Shoes & Footwear") | (df['Category'] == "Books & Stationery")]

print("Products in Shoes & Footwear OR Books & Stationery:")
print(filtered_products.head())
```

```python
pre_order_products = df[df['Availability'] == "pre_order"]

print("Products with Availability = pre_order:")
print(pre_order_products.head())
```

```python
smart_products = df[df['Name'].str.startswith("Smart")]

print("Products whose Name starts with 'Smart':")
print(smart_products.head())
```

# Sorting

```python
sorted_products = df.sort_values(by="Price", ascending=True)

print("Products sorted by Price (ascending):")
print(sorted_products.head(10))   
```

```python
sorted_stock = df.sort_values(by="Stock", ascending=False)

print("Products sorted by Stock (descending):")
print(sorted_stock.head(10))
```

```python
sorted_category_price = df.sort_values(by=["Category", "Price"], ascending=[True, False])

print("Products sorted by Category and then by Price (descending):")
print(sorted_category_price.head(15))   
```

# Renaming & Removing

```python
df = df.rename(columns={
    "Internal ID": "internal_id",
    "EAN": "product_code",
    "Price": "unit_price"
})

print("Columns renamed successfully:")
print(df.head())
```

```python
df.columns = [col.lower().replace(" ", "_") for col in df.columns]

print("Updated column names:")
print(df.columns)
```

# Descriptive Statistics

```python
average_price = df["Price"].mean()

print("Average Price of Products:", average_price)
```

```python
max_price = df["Price"].max()
min_price = df["Price"].min()

print("Maximum Price:", max_price)
print("Minimum Price:", min_price)

print("\nProduct(s) with Maximum Price:")
print(df[df["Price"] == max_price])

print("\nProduct(s) with Minimum Price:")
print(df[df["Price"] == min_price])
```
